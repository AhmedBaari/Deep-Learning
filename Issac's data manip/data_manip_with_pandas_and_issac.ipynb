### Slicing

just like lists

#### Subsetting indexed df using loc, iloc, etc.
1. **df.loc[row1:row2]** : both _row1_ and _row2_ are included. Tuples for accessing inner index. years of date in _yyyy-mm-dd_ format can be used to slice without the month and dates
2. **df.loc[rowslicing,colslicing]**
3. **df.iloc[start:end,:]** : the _end_ index is exclusive here

````python
df.loc[:,"Name":"Ticket"]
````

### Indexing

#### INDEXING Functions, attributes, methods
1. **df.set_index(col)** : to set col as the index for the rows in lieu of 0,1,2,... This helps in calling records using .loc[]. These index values dont need to be unique and there can be multiple columns set as indices where each one is under another in an hierarchical model(called multi-level or hierarchical indexing). multi-level indexed columns are subsetted using df.loc[(a,b),(c,d)] when you to access the second level of the indices. The first level can be subsetted as df.loc[a,c] where a,c are values under upper index and b,d are values under lower index.
2. **df.reset_index(drop=True/False)** : resets index of the obs to 0,1,2,... _drop_ argument drops the set index column while resetting
3. **df.sort_index(level = [col,col2], ascending = True/False)** : sort indices with set level of sorting and sorting order

One-hot encoding (OHE) is a machine learning technique that converts categorical data to numerical values.
