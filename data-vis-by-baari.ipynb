# Plotting in Real World Data

For this session, I had downloaded the world population data from the World Bank. 
https://data.worldbank.org/indicator/SP.POP.TOTL?locations=IN

**GOAL: Plot the population of India over the years.**

### Step 1: Load the data

```python
df = pd.read_csv("./data/API_SP.POP.TOTL_DS2_en_csv_v2_85.csv")
df.head() # To see the first 5 rows
```

### Step 2: Extract India's Data

```python
india_data = df[df["Country Name"] ==  "India"]
india_data
```

### Step 3: Choose the relevant columns, drop the rest

```python
india_data = india_data.drop(["Country Name", "Country Code", "Indicator Name", "Indicator Code"], axis=1)
india_data
```

### Step 4: Prepare the data to be plotted

```python
india_data_t = india_data.transpose()
india_data_t
```

```python
india_data_t = india_data_t.rename(columns={109: "Population"})
india_data_t
```

```python
india_data_t = india_data_t.drop("Unnamed: 67")
```

### Step 5: Plot the data

```python
years = india_data_t.index
populations = india_data_t["Population"]

plt.plot(years, populations)
plt.show()
```

Oh no! The x axis is so clogged up with all the years. Let's fix that.

```python
type(years[0])
```

```python
years = years.astype(int)
```

### Ticks

**What are ticks?**  
Ticks are the values used to show specific points on the coordinate axis.   
By default, Matplotlib will use the values of the data points as ticks on the x-axis and y-axis.   
You can customize the ticks using the xticks() and yticks() functions.  

```python
# np.arange() returns evenly spaced values within a given interval. 
np.arange(min(years), max(years)+1, 10)
```

```python
plt.plot(years, populations)

# Set the x-ticks to be every 10 years
plt.xticks(np.arange(min(years), max(years)+1, 10)) 

plt.show()
```

Now, notice that the years are in multiples of 1e9. But when showing the report, we want to show it in a way understandable to everyone. So let's use the same `ticks` concept to show the years in multiples of billions.

```python
billion = 1000000000
min(populations), max(populations)
```

```python
plt.plot(years, populations)
plt.xticks(np.arange(min(years), max(years)+1, 10))
plt.yticks([
    0.25 * billion,
    0.5 * billion,
    0.75 * billion,
    1 * billion,
    1.25 * billion,
    1.5 * billion,
], [
    "0.25B",
    "0.5B",
    "0.75B",
    "1B",
    "1.25B",
    "1.5B",
])
plt.show()
```

### Adding title and labels

Now, let's add a title and labels to the plot.

```python
plt.plot(years, populations)
plt.xticks(
    [ 1960, 1970, 1980, 1990, 2000, 2010, 2020]
)
plt.yticks([
    0.25 * billion,
    0.5 * billion,
    0.75 * billion,
    1 * billion,
    1.25 * billion,
    1.5 * billion,
], [
    "0.25B",
    "0.5B",
    "0.75B",
    "1B",
    "1.25B",
    "1.5B",
])


plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India')


plt.show()
```

# Correlation Matrix

A correlation matrix is used to investigate the dependence between multiple variables at the same time.  
The result is a table containing the correlation coefficients between each variable and the others.   
This is ideal for finding relationships between data points.

```python
df.head()
```

```python
df_clean = df.drop(columns=['country', 'year','cont'])
df.columns
```

```python
import seaborn as sns

# Create a correlation matrix
correlation_matrix = df_clean.corr()

correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```
